#### schat_6label, lchat_6label 합치기

#### 인덱스 재정렬

```python
schat = pd.read_csv('./data/schat_6label.csv', index_col=0)
schat
```

```python
lchat = pd.read_csv('./data/lchat_6label.csv', index_col=0)
lchat
```

```python
schat.reset_index(drop=True, inplace=True)
```

```python
schat
```

```python
lchat.reset_index(drop=True, inplace=True)
```

```python
lchat
```

```python
chat = pd.concat([schat, lchat], ignore_index=True)
chat
```

```python
chat.groupby('Emotion').count()
```

```python
chat.to_csv('./data/chat_6label.csv', index=True, encoding='utf-8')
```
